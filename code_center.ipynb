```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
def centroid(xs,xf,ys,yf,car_img):
    x = (xf - xs)/2 # (600 - 400)/2 = 100
    x = x+xs # 100 + 400 = 500
    y = (yf - ys)/2 #(300 - 100)/2 = 100
    y = y + ys # 100 + 100 = 200
    
    img_car = cv.circle(car_img,(int(x),int(y)),9,(255,0,0),-6)
    return img_car
```

```python
class Square():
    
    def __init__(self,cx,cy,delt_x,delt_y,img):
        self.cx = cx
        self.cy = cy
        self.delt_x = delt_x
        self.delt_y = delt_y
        self.img  =  img
        
        
    def square_coord(self):
        xs = self.cx - (self.delt_x/2)
        xf = xs + self.delt_x
        ys = self.cy - (self.delt_y/2)
        yf = ys + self.delt_y
        coord =[int(xs),int(xf),int(ys),int(yf)]
        return coord
    
    def create_square(self):
        cv.imread(self.img,2)
        xs,xf,ys,yf = self.square_coord()
        img_car = cv.rectangle(cv.imread(self.img,0),(xs,ys),(xf,yf),(255,255,255),10)
        return img_car
    
    def cut_img(self):
        
        xs,xf,ys,yf = self.square_coord()
        return cv.imread(self.img,1)[ys:yf,xs:xf]
    
    def save_new_image(self,local):
        cut = self.cut_img()
        return cv.imwrite(local,cut)
    
class Look(Square):
    
```
